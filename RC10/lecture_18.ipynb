#### Box Plot

````python
# compare performance for an exam across two semesters

# first read the scores into lists
# scores are out of 60
s22_scores = []
f22_scores = []
for line in open('exam2_s22.txt'):
    s22_scores.append(int(line))
for line in open('exam2_f22.txt'):
    f22_scores.append(int(line))
````

**Minimal box plot**

````python
plt.boxplot(s22_scores)
plt.show()
````

**In a box plot**:
- The box is plotted at x=1
- The center line is at the median value (50th percentile, or 2nd quartile)
- The lower and upper ends of the box are at the 25th and 75th percentile (1st and 3rd quartiles)
- The value band enclosed by the box is called the inter-quartile range (IQR)
- The half lines at top and bottom of the plot are called *whiskers*. The top whisker is at 3Q + 1.5 \* IQR (or max value, if it is more than max), and the bottom whisker is at 1Q - 1.5 \* IQR (or min value, if it is less than min)

````python
print('min:',min(s22_scores))
print('max:',max(s22_scores))
print('median:',np.median(s22_scores))
q1 = np.percentile(s22_scores,25)
q3 = np.percentile(s22_scores,75)
print('25th percentile (1Q):',q1)
print('75th percentile (3Q):',q3)
iqr = q3-q1
print('IQR:',(q3-q1))
bot = q1 - 1.5*iqr
top = q3 + 1.5*iqr
print('bot:',bot,', top:', top)
print('bottom whisker: ',max(min(s22_scores),bot))
print('top whisker: ',min(max(s22_scores),top))
````

**In the minimal plot, the distribution of values is not apparent since the actual values plot is a line that goes over all the points. To get a granular view of the distribution use a scatter plot, and then a box over it.**

````python
plt.scatter([1]*194, s22_scores)  # x value is 1 for all S19 scores, since box is plotted at x=1
plt.boxplot(s22_scores)
plt.show()
````

````python
IRIS_URL = 'http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
columns = ['sepal_length','sepal_width','petal_length','petal_width','iris-type']
iris = pd.read_csv(IRIS_URL,names=columns)
````

````python
iris.describe()
````

````python
iris['iris-type'].value_counts()
````

**Plot sepal width vs sepal length for each of the iris types<br>
Extract subset DataFrames for each of the iris types, and plot directly off DataFrame**

````python
iris_types = iris['iris-type'].unique()
colors = ['blue','red','green']
for iris_type,color in zip(iris_types,colors):
    df = iris[iris['iris-type'] == iris_type]    
    df.plot('sepal_width','sepal_length',kind='scatter',color=color)
````

**Do all of them in the same plot, for easy comparison**

````python
fig, ax = plt.subplots(1,1)
fig.set_size_inches(5,5)
for iris_type,color in zip(iris_types,colors):
    df = iris[iris['iris-type'] == iris_type]    
    df.plot('sepal_width','sepal_length',ax=ax,kind='scatter',color=color,label=iris_type[5:])
# ax.legend(loc='center right')
````

**Repeat for sepal length vs petal length and petal width<br>
Share the y axis since they are all sepal length**

````python
fig, axes = plt.subplots(1,3,sharey=True)
fig.set_size_inches(15,5)
axsubs = [0,1,2]
for x,axsub in zip(['sepal_width','petal_length','petal_width'],axsubs):
    for iris_type,color in zip(iris_types,colors):  
        df = iris[iris['iris-type'] == iris_type]    
        df.plot(x,'sepal_length',ax=axes[axsub],kind='scatter',color=color,label=iris_type[5:])

# remove legends for the first and last plots
axes[0].get_legend().remove()
axes[2].get_legend().remove()
````

````python
housing['income_cat'] = pd.cut(housing['median_income'],
                               bins=[0., 1.5, 3.0, 4.5, 6.0, np.inf],
                               labels=[1,2,3,4,5])
````
